```python
import yaml
import pandas as pd
import numpy as np
import time

interval: str = "1min"

time_series_path: str = f"../data/twelvedata/time series ({interval})/AAPL_time_series.csv"
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

time_series_df: pd.DataFrame = pd.read_csv(time_series_path)
time_series_df['min_last_5'] = time_series_df['close'].shift(1).rolling(window=5).max()
time_series_df['close>30close'] = time_series_df['close'] > time_series_df['min_last_5']
time_series_df
```

```python
print(time_series_df['close>30close'].value_counts())
```

```python
state_column = []
earning_column = []

state = 'idle'
hold_counter = 0

for value in time_series_df.iterrows():
    print(value[2])
    if state == 'idle':
        if value:
            state_column.append('Start')
            earning_column.append(float())
            state = 'hold'
            hold_counter = 5
        else:
            state_column.append(None)
    elif state == 'hold':  # Hold phase
        if hold_counter > 0:
            state_column.append('Hold')
            hold_counter -= 1
        else:
            state_column.append('Stop')
            state = 'idle'
    else:
        state_column.append(None)

time_series_df['pattern_col'] = state_column
time_series_df
```
